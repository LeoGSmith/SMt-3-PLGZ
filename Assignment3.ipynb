# Problem 1: Properties of the van der Waals gas by computation

## 1. Write a program to compute and plot V as a function of P at constant T for the van der Waals gas.

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N = 10
k = 1
a = 0.1
b = 0.1
Tc = 8*a/(27*b*k)
print('Tc is', Tc)
T = 0.285

V = np.linspace(1.7, 10, 200)
P = (N*k*T/(V-b*N))-N**2*a/(V**2)

plt.figure()
plt.plot(P, V, '-^r', label = 'T = 0.285')
plt.xlabel('P(a.u.)')
plt.ylabel('V(a.u.)')
plt.title('problem1-1')
plt.legend()

plt.show()
```

## 2. Modify your program to include a comparison curve for the ideal gas law at the same temperature.

```python
N = 10
k = 1
a = 0.1
b = 0.1
Tc = 8*a/(27*b*k)
print('Tc is', Tc)
T = 0.285

V = np.linspace(1.7, 10, 200)
P = (N*k*T/(V-b*N))-N**2*a/(V**2)

Pi = np.linspace(0.2, 0.6, 100)
Vi = N*k*T/Pi

plt.figure()
plt.plot(P, V, '-^r', label = 'Van der waals gas')
plt.plot(Pi, Vi, '-*b', label = 'ideal gas')
plt.xlabel('P(a.u.)')
plt.ylabel('V(a.u.)')
plt.title('problem1-2 at T = 0.285')
plt.legend()

plt.show()
```

## 3. Make plots for values of T that are at, above, and below Tc

```python
N = 10
k = 1
a = 0.1
b = 0.1
Tc = 8*a/(27*b*k)
print('Tc is', Tc)
T1 = 0.285
T2 = Tc
T3 = 0.35

V = np.linspace(1.7, 10, 200)
P1 = (N*k*T1/(V-b*N))-N**2*a/(V**2)
P2 = (N*k*T2/(V-b*N))-N**2*a/(V**2)
P3 = (N*k*T3/(V-b*N))-N**2*a/(V**2)

plt.figure()
plt.plot(P1, V, '-^r', label = 'T = 0.285')
plt.plot(P2, V, '-^b', label = 'T = Tc')
plt.plot(P3, V, '-^g', label = 'T = 0.35')

plt.xlabel('P(a.u.)')
plt.ylabel('V(a.u.)')
plt.title('problem1-3')
plt.legend()

plt.show()
```
